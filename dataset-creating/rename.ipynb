```python
import os 

def rename_files(directory, real_command, labels=False):
    index = 0
    images = os.listdir(directory)
    images = sorted(images, key=lambda x: int(x.split('_')[1].split('.')[0]))
    for image_file in images:
        command, number = image_file.split('_')
        print(index, number)
        if labels:
            os.rename(f'{directory}/{image_file}', f'{directory}/{real_command}_{index}.jpg.csv')
        else:
            os.rename(f'{directory}/{image_file}', f'{directory}/{real_command}_{index}.jpg')
        index += 1
```

```python
for test in ['train', 'test']:
    for command in ['up', 'down', 'left', 'right', 'open']:
        rename_files(f'./data/{test}/images/{command}/', command)
        rename_files(f'./data/{test}/labels/{command}/', command, labels=True)
```
